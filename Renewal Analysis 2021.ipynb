#### Outlet Data By October 2021

```python
# import outlet data by oct 2021
df_outlet = pd.read_excel('Outlet.xlsx', usecols=["Nama", "Jenis", "Master", "Status"])
```

```python
df_outlet.head()
```

```python
df_outlet.tail()
```

```python
df_outlet.info()
```

```python
# split SILVER 0,5% to SILVER only
df_outlet["Jenis"] = df_outlet.Jenis.str.split(" ", expand=True)[0]
```

```python
# make masks to filter dataframe
mask_jenis = df_outlet.Jenis.isin(['SILVER', 'REGULAR', 'PRIORITAS', 'PLATINUM', 'GOLD'])
mask_master = df_outlet.Master == "AGPG PUSAT"
```

```python
# create new dataframe within jadetabek
df_outlet_jadetabek = df_outlet[mask_jenis & mask_master]
```

```python
df_outlet_jadetabek.set_index("Nama", inplace= True)
```

```python
df_outlet_jadetabek.head()
```

```python
df_outlet_jadetabek.info()
```

```python
#merge df_outlet_jadetabek and df_ayam_aug_to_oct
pd.merge(df_outlet_jadetabek, df_ayam_aug_to_oct, how="left", left_index= True, right_index=True)
```

```python
#assign the merged as new dataframe
df_jadetabek_ayam = pd.merge(df_outlet_jadetabek, df_ayam_aug_to_oct, how="left", left_index= True, right_index=True).fillna(0)
```

```python
#convert parting 5 to parting 4
df_jadetabek_ayam["AYAM PARTING 5"] = df_jadetabek_ayam["AYAM PARTING 5"].mul(5).div(4)
```

```python
# create new column to calculate the average of Ayam Parting 4
df_jadetabek_ayam["average_parting4_perday"] = df_jadetabek_ayam["AYAM PARTING 4"].add(df_jadetabek_ayam["AYAM PARTING 5"]).div(df_jadetabek_ayam.total_order_last_3month).fillna(0)
```

```python
df_jadetabek_ayam["min_realistic_target"] = df_jadetabek_ayam["AYAM PARTING 4"].add(df_jadetabek_ayam["AYAM PARTING 5"]).div(3)
```

```python
df_jadetabek_ayam.rename(columns={"AYAM PARTING 5": "PARTING_5(converted_to_4)",
                                  "AYAM CRISPY (PARTING 9)": "PARTING_9",
                                  "AYAM FILLET (per 10 pcs)": "FILLET",
                                  "AYAM PARTING 4": "PARTING_4",
                                  "Jenis": "Category_Outlet",
                                 "Status": "Status_Web_Order"}, inplace= True)
```

```python
#merge df_jadetabek_ayam and df_targetnov dataframe
pd.merge(df_jadetabek_ayam, df_targetnov, how="left", left_index= True, right_index= True)
```

```python
# assign merged df as new dataframe
df_target_status_evaluation = pd.merge(df_jadetabek_ayam, df_targetnov, how="left", left_index= True, right_index= True)
```

```python
df_target_status_evaluation.Status.fillna("Crosscheck", inplace= True)
df_target_status_evaluation["Target Bulan November 2021"].fillna(0, inplace= True)
```

```python
df_target_status_evaluation.rename(columns={"Status": "Status_by_target_nov"}, inplace=True)
```

```python
df_target_status_evaluation.head()
```

# continue here

```python
# status checking (status_web_order vs status)
# sales target checking (min_realistic vs Target Bulan November 2021)
```

```python
#web order status checking
df_target_status_evaluation["web_order_check"] = df_target_status_evaluation.apply(lambda row: "Normal" if row.Status_Web_Order == "Aktif" and row.Status_by_target_nov == "Aktif" and row.min_realistic_target > 0 and row["Target Bulan November 2021"] > 0
                                 else "Recheck again target" if row.Status_Web_Order == "Aktif" and row.min_realistic_target > 0 and row["Target Bulan November 2021"] == 0
                                                                                   else "Recheck Web Order",axis=1)
```

```python
# target checking
df_target_status_evaluation["target_percentage"] = df_target_status_evaluation.apply(lambda row: 0 if row.min_realistic_target == 0
                                                                                     else (row["Target Bulan November 2021"]/row.min_realistic_target)-1, axis=1)
```

```python
df_target_status_evaluation.head()
```

```python
df_target_status_evaluation["target_status"] = df_target_status_evaluation.apply(lambda row: "Define new target" if row.target_percentage == -1 
                                                                                else "Close/Recheck again" if row.target_percentage == 0
                                                                                 else "Downturn" if row.target_percentage < 0
                                                                                 else "Realistic" if row.target_percentage < 0.3
                                                                                else "Over optimistic"if row.target_percentage >= 0.3
                                                                                 else "Close/Recheck again", axis=1)
```
